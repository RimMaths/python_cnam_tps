Question 2: 
Dans un repère Oxy, la liste des points ainsi que ses valeurs d'abscises et coordonnées sont stockés
dans le fichier position.txt
1. Ecrivez une fonction distance qui prend 2 paramètres étant deux points et qui retourne la distance
entre ces deux points
Ex : distance(A,B) --> 5
2. Ecrivez une fonction plusProche permettant de chercher le point le plus proche en se basant sur sa
distance avec les autres points.
Ex : plusProche(A) --> K
3. Ecrivez une fonction estRectangle qui prend 4 paramètres (4 points) et renvoie si on pourra faire un
rectangle constitué de ces 4 points*
Ex : estRectangle(A, L, M, K) --> TrueG

```python
#ouvertir du fichier position juste en lecture 
f = open("position.txt","r")
#je veux voir ce qui a dedans
for ligne in f:
    print(ligne)

f.close()
```

```python
#Là je veux nettoyer la donnée du fichier position que j'ai 
points = {}

with open("position.txt", "r") as f:
    for ligne in f:
        ligne = ligne.strip()

        # Si ligne vide → on saute
        if not ligne:
            continue

        nom, coord = ligne.split(';')   # Séparer nom et "(x,y)"

        coord = coord.strip()
        coord = coord.strip('()')       # enlever ( )

        x_str, y_str = coord.split(',')
        x, y = int(x_str), int(y_str)

        points[nom] = (x, y)

print("Points extraits :", points)



# Étape 2 - Fonction distance
import math
def distance(p1, p2):
    x1, y1 = points[p1]
    x2, y2 = points[p2]
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Exemple
print("Distance ||AB|| :", distance('A', 'B'))


# Étape 3 - Fonction plusProche
def plusProche(p):
    distances = {nom: distance(p, nom) for nom in points if nom != p}
    return min(distances, key=distances.get)

# Exemple
print("Point plus proche de A :", plusProche('A'))


# Étape 4 - Fonction estRectangle (vérification simple)
def estRectangle(a, b, c, d):
    dists = [
        # côtés
        distance(a, b), distance(b, c), distance(c, d), distance(d, a),  
        # diagonales
        distance(a, c), distance(b, d)                                   
        ]
    cote1, cote2, cote3, cote4, diag1, diag2 = dists

    # Vérifier que côtés opposés sont égaux et diagonales égales
    return (math.isclose(cote1, cote3) and
            math.isclose(cote2, cote4) and
            math.isclose(diag1, diag2))

# Exemple
print("Est-ce qu(A, L, M, K) forment un rectangle ?", estRectangle('A', 'L', 'M', 'K'))
```

Question 3 : 
Soit un fichier intitulé concours.txt qui comporte les enregistrements relatifs aux candidats d’un
concours. Chaque enregistrement est composé de : PRENOM, NOM, NOTE, DECISION : (type contenant
les identificateurs suivants : admis, refusé, ajourné), et séparé par point-virgule ( ;).
1. Définir la fonction saisir() qui permet de remplir à nouveau les données relatives aux candidats dans
le fichiers concours.txt
2. Définir la fonction admis() qui permet de créer le fichier admis.txt comportant les données relatives
aux candidats admis.
3. Mettre à jour le fichier admis.txt en triant les candidats par ordre croissant de notes obtenues.
4. Afin de sélectionner en priorité les candidats admis et ayant une note supérieure à 16, écrivez une
fonction priorite() qui filtra à partir du fichier admis.txt et qui produira une nouveau fichier prorité.txt.
Une ligne du nouveau fichier comprend le nom, prénom et la note d'un candidat séparés par pointvirgule (;)
5. Définir la fonction statistiques() qui permet de retourner le pourcentage des candidats en fonction
de la décision (admis, refusé et ajourné).
6. Définir la fonction supprimer() qui supprimera du fichier concours.txt les candidats qui ont été
refusés

```python
#fonction saisir 
def saisir():
    """ Remplir le fichier concours.txt avec les candidats  saisis au clavier """
    with open("concours.txt", "w") as f:
        n = int(input("Combien de candidats voulez-vous saisir ? "))
        for _ in range(n):
            prenom = input("Prénom du candidat : ")
            nom = input("Nom du candidat : ")
            note = float(input("Note du candidat : "))
            decision = input("Décision (admis/refusé/ajourné) : ")
            #construction de ligne dans le bon format que le fichier 
            ligne = f"{prenom};{nom};{note};{decision}\n"
            #Ecriture du fichier 
            f.write(ligne)
        #affichage d'un message de confirmation
        print("concours.txt rempli avec succès ")
        
#appel de la fonction saisir
saisir()
```

```python
#foction admis 
def admis():
    """Crée le fichier admis.txt avec les candidats admis."""
    with open("concours.txt", "r") as f_in, open("admis.txt", "w") as f_out:
        for ligne in f_in:
            ligne = ligne.strip()
            if not ligne:
                continue

            prenom, nom, note_str, decision = ligne.split(";")
            if decision.strip().lower() == "admis":
                f_out.write(ligne + "\n")

    print(">> admis.txt créé.")
#appel de la fonction admis
admis()
```

```python
# 3) Trier admis.txt par ordre croissant des notes
def trier_admis():
    """Trie le fichier admis.txt par note croissante."""
    candidats = []

    # Lire les admis
    with open("admis.txt", "r") as f:
        for ligne in f:
            ligne = ligne.strip()
            if not ligne:
                continue
            prenom, nom, note_str, decision = ligne.split(";")
            note = float(note_str)
            candidats.append((prenom, nom, note, decision))

    # Trier par note (3e élément du tuple)
    candidats.sort(key=lambda c: c[2])
    # Réécrire le fichier trié
    with open("admis.txt", "w") as f:
        for prenom, nom, note, decision in candidats:
            ligne = f"{prenom};{nom};{note};{decision}\n"
            f.write(ligne)
    print(">> admis.txt trié par note.")
#appel de la fonction trier_admis
trier_admis()
```

```python
# 4) Fonction priorite() : admis avec note > 16 dans priorite.txt
def priorite():
    """
    Crée priorite.txt avec les candidats admis ayant une note > 16.
    Format d'une ligne : NOM;PRENOM;NOTE
    """
    with open("admis.txt", "r") as f_in, open("priorite.txt", "w") as f_out:
        for ligne in f_in:
            ligne = ligne.strip()
            if not ligne:
                continue

            prenom, nom, note_str, decision = ligne.split(";")
            note = float(note_str)

            if note > 16:
                # On écrit NOM;PRENOM;NOTE comme demandé
                f_out.write(f"{nom};{prenom};{note}\n")

    print(">> priorite.txt créé.")
#appel de la fonction priorite
priorite()
```

```python
# 5) Fonction statistiques() : pourcentage par décision
def statistiques():
    """
    Retourne un dictionnaire avec le pourcentage de candidats
    admis, refusés et ajournés.
    """
    total = 0
    compteur = {"admis": 0, "refusé": 0, "ajourné": 0}

    with open("concours.txt", "r") as f:
        for ligne in f:
            ligne = ligne.strip()
            if not ligne:
                continue

            prenom, nom, note_str, decision = ligne.split(";")
            decision = decision.strip().lower()
            total += 1

            if decision in compteur:
                compteur[decision] += 1

    if total == 0:
        return {"admis": 0.0, "refusé": 0.0, "ajourné": 0.0}

    # Calcul des pourcentages
    stats = {cle: (compteur[cle] * 100) / total for cle in compteur}

    # Affichage (optionnel)
    print("Statistiques :")
    for cle, pct in stats.items():
        print(f"{cle} : {pct:.2f} %")

    return stats


#appel de la fonction statistiques
statistiques()
```

```python
# 6) Fonction supprimer() : enlever les refusés de concours.txt
def supprimer():
    """Supprime du fichier concours.txt les candidats refusés."""
    lignes_conservees = []

    # Lire tout en mémoire
    with open("concours.txt", "r") as f:
        for ligne in f:
            ligne = ligne.strip()
            if not ligne:
                continue

            prenom, nom, note_str, decision = ligne.split(";")
            decision = decision.strip().lower()

            if decision != "refusé":
                lignes_conservees.append(f"{prenom};{nom};{note_str};{decision}")

    # Réécrire le fichier sans les refusés
    with open("concours.txt", "w") as f:
        for ligne in lignes_conservees:
            f.write(ligne + "\n")

    print(">> candidats refusés supprimés de concours.txt.")
    
#appel de la fonction supprimer
supprimer()
```
