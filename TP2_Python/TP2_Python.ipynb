Question 2) : nombre d'amstong (371 = 3^3+7^3+1)

```python
amstrong = []
for nbr in range(1,2000):
    s=0
    tmp = nbr
    while tmp > 0 :
        #d fait référence de maniere à prendre les dizaines , unité centaines 
        d = tmp % 10
        s+= d**3
        tmp//=10
    if nbr == s:
        amstrong.append(nbr)
print(amstrong)
```

Question 3) Générez la liste de tous les bigrammes (chaine de deux lettres , par exemple AA , AC , NG etc ..)

```python
import string 
#list des alphabets déjà éxitants
alphabet = string.ascii_uppercase 
bigrammes = []
for lettre1 in alphabet:
    for lettre2 in alphabet :
        bigrammes.append(lettre1 + lettre2)
print(bigrammes)
print("le nombre de possibilitées", len(bigrammes))
```

Question 4 : Écrire un algorithme en langage Python qui permet de renvoyer une liste formée des mots communs
à deux textes. Exemple si T1 = "Python est un langage de programmation" et T2 = "Python est orienté
objet" , le programme doit renvoyer la liste des mots communs L = ['Python' , 'est' ].


```python
T1 = "Python est un langage de programmation"
T2 = "Python est orienté objet"
mots_T1 = T1.split()
#["Python", "est", "un", "langage", "de", "programmation"]
mots_T2 = T2.split()
communs = []
for mot in mots_T1:
    if mot in mots_T2:
        if mot not in communs: 
            communs.append(mot)
print(communs)
```

Question 5 :
Écrire un programme Python qui remplace les caractères d’index impaires d’une chaîne donnée
par ‘#’. Exemple : si s = “Python”, l’algorithme renvoie la chaîne : “P#t#o#”


```python
s = "Python"
lists = list(s)
for i in range(len(lists)):
    if i % 2 !=0:
        lists[i]= '#'
resultat = ''.join(lists)
print(resultat)

```

Question 6 :
Ecrire une fonction nommée compte_mots_ligne() qui accepte une chaine de caractères et qui renvoie
l'occurence des mots contenus dans la chaine sous forme de dictionnaire.

```python
phrase = " "
resultat = [] 
def compte_mots_ligne(phrase): 
    mots = phrase.split()
    
    # Créer un dictionnaire pour stocker les occurrences
    occurrences = {}
    
    # Compter les occurrences de chaque mot
    for mot in mots:
        mot = mot.lower()  # pour ignorer la casse
        if mot in occurrences:
            occurrences[mot] += 1
        else:
            occurrences[mot] = 1
    
    return occurrences
# Saisie utilisateur
phrase = input("Entrez une phrase : ")
resultat = compte_mots_ligne(phrase)

# Affichage du résultat
print("Occurrences des mots :", resultat)
```

Question 7 :
Etant données deux listes d’entiers, on souhaite écrire un prédicat qui teste si les deux listes de
longueur respective n et m, n > m sont équivalentes. On dira que deux listes sont équivalentes si elles
contiennent les mêmes éléments, peu importe l’ordre et le nombre d’occurrences. Par exemple :
[1; 9; 6; 1; 1; 5] et [ 5; 6; 1; 9; 5] sont équivalentes, [ 5; 6; 1; 9; 5] et [ 6; 1; 9; 1] ne sont pas équivalentes
(la valeur 5 n’apparaît pas). Ecrire un algorithme qui réalise cette tâche.


```python
def sont_equivalentes(liste1, liste2):
    # On identifie la plus grande et la plus petite liste
    grande = liste1 if len(liste1) > len(liste2) else liste2
    petite = liste2 if grande == liste1 else liste1

    # On transforme la grande liste en ensemble (éléments uniques)
    ensemble_grande = set(grande)

    # On vérifie que chaque élément de la petite liste est dans la grande
    for element in set(petite):
        if element not in ensemble_grande:
            return False
    return True
print(sont_equivalentes([1, 9, 6, 1, 1, 5], [5, 6, 1, 9]))
print(sont_equivalentes([5, 6, 1, 9], [6, 1, 9, 1]))
```
