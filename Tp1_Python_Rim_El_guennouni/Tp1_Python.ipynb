Queston 2 : 
Ecrivez , en utilisant la boucle for , une fonction compte_triangle avec un parametre n qui compte le nombre d'étoiles dans un triangle de n lignes ; résultat : compte_triangle(5)-> résultat attendu : 15 

```python
def compte_triangle(n): 
    compteur = 0 
    #on incrémente
    for i in range(1 , n+1):
        ligne = ""
        for j in range(1 , i+1 ): 
            ligne += "*"
            compteur += 1
        print(ligne)
    return compteur 
resultat = compte_triangle(5)
print("nombre total d'étoiles",resultat)
```

Question 4 : 
Ecrivez une fonction nb_differences avec deux paramètres s1 et s2 de type chaînes caractères qui
compte le nombre de lettres différentes entre s1 et s2.
Par exemple : pour les chaînes « oiseau » et « bateau », il y a 3 premières lettres différentes
o i s e a u
b a t e a u

```python
def nb_differences(s1, s2):
    compteur = 0
    longueur = min(len(s1) , len(s2))
    for i in range(longueur):
        if s1[i] != s2[i]:
            compteur +=1
    return compteur 
s1 = "oiseau" 
s2 = "bateau"
print(nb_differences(s1, s2))
```

Ecrire un programme qui affiche un sapin de Noël, dont la taille est donnée par l’utilisateur. Exemple
pour une taille de 6 :
      *
     ***
    *****
   *******
  *********
 ***********

```python
def triangle (L) : 
    for i in range(1 , L+1):
        #tracer les éspaces 
        for j in range(1 , L-i+1):
            print(" " , end="")
        #tracer les étoiles
        for j in range(1, 2*i):
            print("*" , end="")
        print()
L = int(input("Veuillez saisir le nombre de lignes : "))
triangle(L)
```
