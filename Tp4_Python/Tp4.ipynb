```python
# un mot de passe al√®atoire en moins de 8 caractere
#Pr√©parer la liste des caract√®res possibles 
minuscule = "abcdefghijklmnopqrstuvwxyz"
majuscule = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
chiffre = "0123456789"
caracteres_speciaux = "!@#$%^&*()-_=+[]{};:,.<>?/|"

# Fonction pour g√©n√©rer un mot de passe al√©atoire
def password_generator (longueur): 
    
#mot de passe moins de 8 caractere
    if longueur < 8:
        #on force la longueur √† 8
        longueur = 8
        
    tous_les_caracteres = minuscule + majuscule + chiffre + caracteres_speciaux
    mot_de_passe = " "
    for _ in range (longueur):
        mot_de_passe += random.choice(tous_les_caracteres)
        
    return mot_de_passe
mot_de_passe_genere = password_generator(6)
print("Mot de passe g√©n√©r√© :", mot_de_passe_genere)
print("Longueur du mot de passe :", len(mot_de_passe_genere))
```

Question 2 : Rock Paper Scissors

Winning Rules as follows:
Rock vs paper-> paper wins
Rock vs scissor-> Rock wins
paper vs scissor-> scissor wins.

Etude de solution ; 
	       ROCK	             PAPER	                  SCISSOR
ROCK	 Egalite (0)	    Paper gagne (2)	    Rock ( 1) gagne
PAPER	 Paper gagne (1) 	Egalite (0)	        Scissor gagne (2)
SCISSOR	 Rock ( 2 ) gagne 	Scissor gagne (1) 	Egalite (0)
		
           (1 gagne )        (2gagne)               (egal)              
L√©gende :
‚Ä¢	(0) ‚Üí √©galit√©
‚Ä¢	(1) ‚Üí joueur 1 gagne
‚Ä¢	(2) ‚Üí joueur 2 gagne

Maths : 

	   ROCK	 PAPER	 SCISSOR
ROCK	0	  2	      1
PAPER	1 	  0	      2
SCISSOR	2 	  1	      0

Les lignes choix du joueur 1 
Les colonnes choix du joueur 2 



```python
# Rock Paper Scissors - TP4 Q2

# 1. D√©finition des choix possibles et de la matrice des r√®gles
valid_choices = ["rock", "paper", "scissor"]

# Matrice des r√®gles : 
# 0 = √©galit√©, 1 = joueur 1 gagne, 2 = joueur 2 gagne
rules = {
    "rock":   {"rock": 0, "paper": 2, "scissor": 1},
    "paper":  {"rock": 1, "paper": 0, "scissor": 2},
    "scissor":{"rock": 2, "paper": 1, "scissor": 0}
}

# 2. Fonction pour d√©terminer le gagnant
def get_winner(choice1, choice2):
    """
    Retourne :
    0 si √©galit√©
    1 si le joueur 1 gagne
    2 si le joueur 2 gagne
    """
    return rules[choice1][choice2]

# 3. Fonction pour demander un choix √† un joueur
def ask_choice(player_number):
    """
    Demande le choix d'un joueur (1 ou 2)
    et v√©rifie qu'il est valide.
    """
    choice = input(f"Joueur {player_number}, fais ton choix (rock/paper/scissor) : ").lower()
    
    while choice not in valid_choices:
        print("Choix invalide. R√©essaie.")
        choice = input(f"Joueur {player_number}, fais ton choix (rock/paper/scissor) : ").lower()
    return choice
# 4. Fonction pour jouer un seul round
def play_one_round():
    print("\n--- Nouveau Round ---")
    
    choice1 = ask_choice(1)
    choice2 = ask_choice(2)

    print(f"\nJoueur 1 a choisi : {choice1}")
    print(f"Joueur 2 a choisi : {choice2}")
    
    result = get_winner(choice1, choice2)
    
    if result == 0:
        print("R√©sultat : √âgalit√© üòÖ")
    elif result == 1:
        print("R√©sultat : Bravo Joueur 1 ! üèÜ")
    else:
        print("R√©sultat : Bravo Joueur 2 ! üèÜ")
# 5. Fonction principale pour g√©rer le jeu
def rock_paper_scissors_game():
    print("Bienvenue dans Rock Paper Scissors ! üéÆ")
    
    while True:
        play_one_round()
        
        again = input("\nVoulez-vous rejouer ? (o/n) : ").lower()
        if again != "o":
            print("Merci d'avoir jou√© ! üëã")
            break

# 6. Lancer le jeu
rock_paper_scissors_game()
```

Question 3 : Jeu de num√©ro

Analyse de la probl√®matique : 

Un espace de recherche fini :

un nombre de 4 chiffres 
N = d1d2d3d4 
o√π chaque di appartient {0,1,3,4,...,9}
Le joueur propose un nombreb ; G = g1g2g3g4
pour chaque position i appartient {1,2,3,4}:
Si gi = di ‚Üí ‚≠ê √©toile
Sinon ‚Üí üî® marteau

On compte donc : 
    √©toiles = |{i| gi = di}|
    marteaux = 4 - √©toiles 
L'espace de l'ensemble : 
S = {0000,0001, ... ,9999}
|S|= 10^4 = 10000

```python
import random

# 1) G√©n√©rer un nombre secret √† 4 chiffres (ex: "0234", "9876")
def generate_secret():
    # entier entre 0 et 9999
    n = random.randint(0, 9999) 
    # formatte sur 4 chiffres, avec des z√©ros devant si besoin  
    secret = f"{n:04d}"           
    return secret

# 2) Calculer le nombre d'√©toiles et de marteaux
def compute_score(secret, guess):
    """
    secret et guess sont des cha√Ænes de 4 caract√®res,
    chacun repr√©sentant un chiffre.
    
    Retourne (etoiles, marteaux)
    etoiles  = nombre de chiffres corrects √† la bonne position
    marteaux = 4 - etoiles
    """
    etoiles = 0
    for i in range(4):              # positions 0,1,2,3
        if guess[i] == secret[i]:   # m√™me chiffre et m√™me position
            etoiles += 1
    marteaux = 4 - etoiles
    return etoiles, marteaux

# 3) Jeu principal
def play_number_game():
    print("===== Jeu de num√©ro : √âtoiles & Marteaux =====")
    secret = generate_secret()
    attempts = 0
    
    
    # print("(DEBUG) Secret =", secret)
    
    while True:
        guess = input("Propose un nombre √† 4 chiffres (ex : 1234) : ")
        
        # V√©rifier que l'entr√©e est bien 4 chiffres
        if len(guess) != 4 or not guess.isdigit():
            print("Entr√©e invalide. Tu dois saisir exactement 4 chiffres (ex : 0073).")
            continue
        
        attempts += 1
        
        # Calculer √©toiles et marteaux
        etoiles, marteaux = compute_score(secret, guess)
        print(f"‚Üí R√©sultat : {etoiles} √©toile(s), {marteaux} marteau(x).")
        
        # Condition de fin : 4 √©toiles = tout est correct
        if etoiles == 4:
            print(f"üéâ Bravo ! Tu as trouv√© le nombre {secret} en {attempts} essai(s).")
            break

# 4) Lancer le jeu
play_number_game()
```
